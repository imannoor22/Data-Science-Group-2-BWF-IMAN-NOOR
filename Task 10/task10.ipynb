# **Introduction to Numpy**
- NumPy is a Python library used for working with arrays.
- It also has functions for working in domain of linear algebra, fourier transform, and matrices. 
- NumPy stands for **`Numerical Python`**.

## **Why Use NumPy?**
- NumPy aims to provide an array object that is up to 50x faster than traditional Python lists.
- The array object in NumPy is called `ndarray`, it provides a lot of supporting functions that make working with ndarray very easy.
- Arrays are very frequently used in data science, where speed and resources are very important.

## **Why is NumPy Faster Than Lists?**
- NumPy arrays are stored at one continuous place in memory unlike lists, so processes can access and manipulate them very efficiently.
- This behavior is called `locality of reference` in computer science.
- This is the main reason why NumPy is faster than lists. Also it is optimized to work with latest CPU architectures.

```python
mean = 4.11
std_dev = 1.37
x = 3.00

# Calculate the cumulative probability
probability = norm.cdf(x, mean, std_dev)

percentage = probability * 100

print(f"The percentage of customers who will spend less than $3.00 on concessions is approximately {percentage:.2f}%.")
```

```python
mean = 4.11
std_dev = 1.37

# Percentile (top 87th percentile)
percentile = 0.87

# Calculate the value corresponding to the top 87th percentile
value = norm.ppf(percentile, mean, std_dev)

print(f"The spending amount corresponding to the top 87th percentile is approximately ${value:.2f}.")
```

```python
mean = 4300
std_dev = 750

# Bounds
upper_bound = 4200
lower_bound = 2500

# Calculate the cumulative probability
upper_prob = norm.cdf(upper_bound, mean, std_dev)
lower_prob = norm.cdf(lower_bound, mean, std_dev)

probability = upper_prob - lower_prob

print(f"The probability between 2500 and 4200 acres is approximately {probability:.4f}.")
```

## **Uniform Distribution**
- Uniform distributions are probability distributions with equally likely outcomes. 
    - In a `discrete` uniform distribution, outcomes are discrete and have the same probability. 
    - In a `continuous` uniform distribution, outcomes are continuous and infinite.
    


```python
a = 0
b = 23
theoretical_mu = (a+b)/2
theoretical_sigma = np.sqrt(((b-a)**2)/12)
print("Theoretical Mean:", theoretical_mu)
print("Theoretical Standard Deviation:", theoretical_sigma)
```

## **As sample size is large so it is better to visualize it**

```python
low_boundary = 0.0
high_boundary = 1.0
uniform_samples = np.random.uniform(low_boundary, high_boundary, 1000)
plt.hist(uniform_samples, bins=30, edgecolor="black", color="#D8BFD8")
plt.title("Uniform Distribution", weight="bold", fontsize=16)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## **Binomial Distribution**
- Binomial distribution is calculated by multiplying the probability of success raised to the power of the number of successes and the probability of failure raised to the power of the difference between the number of successes and the number of trials.



```python
n = 12    # Number of trials
p = 0.5   # Probability of success

prob_7_heads = np.math.factorial(n)/(np.math.factorial(7)*np.math.factorial(n-7))*(p ** 7)*((1-p) ** (n-7))
print("Probability of exactly 7 heads:", prob_7_heads)
```

```python
n = 20
p = 0.5
binomial_samples = np.random.binomial(n, p, 1000)
plt.hist(binomial_samples, edgecolor="black", color="#C1E1C1")
plt.title("Binomial Distribution", weight="bold", fontsize=16)
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

## **Poisson Distribution**
- A Poisson distribution is a discrete probability distribution. It gives the probability of an event happening a certain number of times (k) within a given interval of time or space. 



```python
lam = 5
size = 1000 
poisson_samples = np.random.poisson(lam, size)
plt.hist(poisson_samples, bins=20, edgecolor='black')
plt.title("Poisson Distribution", weight="bold", fontsize=16)
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.show()
```

**13. Simulate rolling a six-sided die 1000 times and count the frequency of each outcome.**

**Represent a die**
- The first thing you will need is to define how many sides your dice will have. Reference: [Dice_Simulation](https://github.com/Ryota-Kawamura/Mathematics-for-Machine-Learning-and-Data-Science-Specialization/blob/main/Course-3/Week-2/C3_W2_Lab_2_Dice_Simulations.ipynb)

```python
n_sides = 6 # no. of sides

# Dice is represented using a numpy array
die = np.array([i for i in range(1, n_sides+1)])
die
```

**Roll the die**
- With your die ready it is time to roll it. For now you will assume that the die is fair, which means the probability of landing on each side is the same (it follows a uniform distribution). To achieve this behaviour you can use the function np.random.choice, which given a numpy array returns one of the entries in it randomnly:

```python
np.random.choice(die)
```

*This is great but if you wanted to roll the die 1000 times you will need to run the cell 1000 times and record each result. Now you need a way to simulate several rolls at the same time. For this you can define the number of rolls you desire and use a list comprehension to roll the die as many times as you like, you can also save every roll in a numpy array:*

```python
n_rolls = 1000

# Saving the result
rolls = np.array([np.random.choice(die) for _ in range(n_rolls)])
rolls
```

*Now you have a convenient way of keeping track of the result of each roll, nice!*

*What is you would like to know the mean and variance of this process. For this you can use numpy's functions np.mean and np.var:*

```python
# Compute mean of 1000 rolls
m = np.mean(rolls)

# Compute variance of 1000 rolls
v = np.var(rolls)

print(f"Mean of rolls: {m:.2f}\nVariance of rolls: {v:.2f}")
```
