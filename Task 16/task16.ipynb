# **Introduction to Matplotlib for Data VisualizationðŸ“Š**

Matplotlib is a powerful library in Python used for creating static, animated, and interactive visualizations. It provides a MATLAB-like interface and is highly customizable, making it suitable for a wide range of plotting needs in data analysis and scientific research.

## Key Features of Matplotlib

- **Plot Types**: Matplotlib supports various plot types such as line plots, scatter plots, bar plots, histograms, pie charts, and more.
  
- **Customization**: Almost every aspect of plots can be customized, including colors, labels, annotations, axis limits, and styles.
  
- **Publication Quality**: Matplotlib is designed to produce publication-quality figures, suitable for inclusion in scientific papers and presentations.
  
- **Integration**: It seamlessly integrates with Jupyter notebooks, allowing interactive exploration of data and quick iteration on visualizations.

## Importing libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Plotting x and y points
- The `plot()` function is used to draw points (markers) in a diagram.
- By default, the plot() function draws a line from point to point.
- The function takes parameters for specifying points in the diagram.
- Parameter 1 is an array containing the points on the x-axis(The x-axis is the horizontal axis.).
- Parameter 2 is an array containing the points on the y-axis(The y-axis is the vertical axis.).


## Draw a line in a diagram from position (0,0) to position (6,250):

```python
x = np.array([0, 6])
y = np.array([0, 250])
plt.plot(x, y)
plt.show()
```

## Draw a line in a diagram from position (1, 3) to position (8, 10):

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10]) 
plt.plot(xpoints, ypoints)
plt.show()
```

## Plotting Without Line
- To plot only the markers, you can use *shortcut string notation* parameter 'o', which means 'rings'.

## Draw two points in the diagram, one at position (1, 3) and one in position (8, 10):

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10]) 
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

## Multiple Points
- You can plot as many points as you like, just make sure you have the same number of points in both axis.

## Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10]) 
plt.plot(xpoints, ypoints)
plt.show()
```

## Default X-Points
- If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints)
plt.show()
```

*The x-points in the example above are [0, 1, 2, 3, 4, 5].*

## Markers
- The keyword argument `marker` is used to emphasize each point with a specified marker:

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='o')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='P') # 'P' for plus sign
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='*')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='D') # 'D' for diamond
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker='|') # '|' for Vline
plt.show()
```

## Format Strings `fmt`
- The shortcut string notation parameter is used to specify the marker.
- This parameter is also called `fmt`, and is written with this syntax:

> `marker|line|color`

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, 'o:c')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, 'o-.g') # '-.' for dashed/dotted line
plt.show()
```

## Marker Size & Marker Color
- The keyword argument `markersize` or the shorter version, `ms` and `markeredgecolor` or the shorter `mec` is used to set the *size* and *color of the edge* of the markers respectively:

```python
ypoints = np.array([3, 5, 9, 10])
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50') # `mfc` for face marker color
plt.show()
```

## Linestyle
- The keyword argument `linestyle`, or shorter `ls`, to change the style of the plotted line:

```python
plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

## Shorter Syntax
The line style can be written in a shorter syntax:

- `linestyle` can be written as `ls`.
- `dotted` can be written as `:`.
- `dashed` can be written as `--`.

```python
plt.plot(ypoints, ls=':')
plt.show()
```

```python
plt.plot(ypoints, ls='--', color = 'r')
plt.show()
```

```python
plt.plot(ypoints, color = 'm', linewidth = '20.1')
plt.show()
```

Many lines can be plotted by simply adding more plt.plot() functions:

## Create Labels and Title for a Plot
- With Pyplot, `xlabel()`, `ylabel()` and `title()` functions can be used to set a label for the x- and y-axis and title respectively.

## Add labels to the x- and y-axis and also title:

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.xlabel("Average Pulse", weight='bold', fontsize=15)
plt.ylabel("Calorie Burnage", weight='bold', fontsize=15)
plt.title("Sports Watch Data", weight='bold', fontsize=25)
plt.plot(x, y)
plt.show()
```

## Set Font Properties for Title and Labels
- The fontdict parameter in `xlabel()`, `ylabel()` and `title()` is used to set font properties for the title and labels.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.xlabel("Average Pulse", fontdict=font2)
plt.ylabel("Calorie Burnage", fontdict=font2)
plt.title("Sports Watch Data", fontdict=font1)
plt.plot(x, y)
plt.show()
```

## Position the Title
- The `loc` parameter in `title()` to position the title.

Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

```python
plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

## Add Grid Lines to a Plot
- With Pyplot, the `grid()` function is used to add grid lines to the plot.

```python
plt.xlabel("Average Pulse", weight='bold', fontsize=15)
plt.ylabel("Calorie Burnage", weight='bold', fontsize=15)
plt.title("Sports Watch Data", weight='bold', fontsize=25)
plt.plot(x, y)

plt.grid()

plt.show()
```

## Display only grid lines for the x-axis:

```python
plt.xlabel("Average Pulse", weight='bold', fontsize=15)
plt.ylabel("Calorie Burnage", weight='bold', fontsize=15)
plt.title("Sports Watch Data", weight='bold', fontsize=25)
plt.plot(x, y)

plt.grid(axis='y')

plt.show()
```

# Titanic Dataset Exploration ðŸš¢

```python
titanic = pd.read_csv('titanic.csv')
df = titanic.head(10)
df.head()
```

## Basic Plots:
- Create a simple line plot.
- Plot multiple lines on the same graph.
- Add markers to a line plot.
- Customize line styles and colors.

```python
# Create a simple line plot.
plt.plot(df['Age'], df['Fare'])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title('Fare by Age', fontdict=font1)
plt.xlabel('Age', fontdict=font2)
plt.ylabel('Fare', fontdict=font2)
plt.show()
```

```python
# Plot multiple lines on the same graph.
plt.plot(df['Age'], df['Fare'], label='Fare')
plt.plot(df['Age'], df['Pclass'], label='Pclass')
plt.xlabel('Age', weight='bold', fontsize=10)
plt.ylabel('Values', weight='bold', fontsize=10)
plt.title('Age vs. Fare and Pclass', weight='bold', fontsize=20)
plt.legend()
plt.show()
```

```python
# Add markers to a line plot.
plt.plot(df['Age'], df['Fare'], marker='o', label='Fare')
plt.xlabel('Age', weight='bold', fontsize=10)
plt.ylabel('Fare', weight='bold', fontsize=10)
plt.title('Age vs. Fare with Markers', weight='bold', fontsize=20)
plt.legend()
plt.show()
```

```python
# Customize line styles and colors.
plt.plot(df['Age'], df['Fare'], linestyle='--', color='r', label='Fare')
plt.plot(df['Age'], df['Pclass'], linestyle='-', color='b', label='Pclass')
plt.xlabel('Age', weight='bold', fontsize=10)
plt.ylabel('Values', weight='bold', fontsize=10)
plt.title('Customized Line Styles and Colors', weight='bold', fontsize=20)
plt.legend()
plt.show()
```

## Scatter Plots:
- Create a basic scatter plot.
- Customize marker styles, sizes, and colors.
- Plot multiple scatter plots on the same graph.
- Add labels and titles to the scatter plot.

```python
# Create a basic scatter plot.
plt.scatter(df['Age'], df['Fare'])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title('Fare by Age', fontdict=font1)
plt.xlabel('Age', fontdict=font2)
plt.ylabel('Fare', fontdict=font2)
plt.show()
```

```python
# Customize marker styles, sizes, and colors.
plt.scatter(df['Age'], df['Fare'], c='r', marker='x', s=100)
plt.xlabel('Age', weight='bold', fontsize=10)
plt.ylabel('Fare', weight='bold', fontsize=10)
plt.title('Customized Scatter Plot', weight='bold', fontsize=20)
plt.show()
```

```python
# Plot multiple scatter plots on the same graph.
plt.scatter(df['Age'], df['Fare'], c='r', label='Fare')
plt.scatter(df['Age'], df['Pclass'], c='b', label='Pclass')
plt.xlabel('Age', weight='bold', fontsize=10)
plt.ylabel('Values', weight='bold', fontsize=10)
plt.title('Age vs. Fare and Pclass Scatter Plot', weight='bold', fontsize=20)
plt.legend()
plt.show()
```

## Bar Charts:
- Create a vertical bar chart.
- Create a horizontal bar chart.
- Customize bar colors and add edge colors.

```python
# Create a vertical bar chart.
plt.figure(figsize=(10, 6))
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
df['Pclass'].value_counts().plot(kind='bar')
plt.xlabel('Passenger Class', fontdict=font2)
plt.ylabel('Frequency', fontdict=font2)
plt.title('Passenger Class Frequency', fontdict=font1)
plt.show()
```

```python
# Create a horizontal bar chart.
plt.figure(figsize=(10, 6))
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
df['Pclass'].value_counts().plot(kind='barh')
plt.xlabel('Passenger Class', fontdict=font2)
plt.ylabel('Frequency', fontdict=font2)
plt.title('Passenger Class Frequency', fontdict=font1)
plt.show()
```

```python
# Customize bar colors and add edge colors
plt.figure(figsize=(10, 6))
df['Pclass'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Passenger Class', weight='bold', fontsize=15)
plt.ylabel('Frequency',  weight='bold', fontsize=15)
plt.title('Passenger Class Frequency with Customized Colors',  weight='bold', fontsize=20)
plt.show()
```

## Pie Charts:
- Create a basic pie chart.
- Customize slice colors and add a legend.
- Explode a slice of the pie chart.
- Add percentage labels to slices.

```python
# Basic pie chart
plt.figure(figsize=(6, 6))
df['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Passenger Class Distribution', weight='bold', fontsize=20)
plt.show()
```

```python
# Customize slice colors and add a legend
colors = ['#A2D9D9', '#005F60']
plt.figure(figsize=(6, 6))
df['Pclass'].value_counts().plot(kind='pie', colors=colors, autopct='%1.1f%%')
plt.title('Passenger Class Distribution with Custom Colors', weight='bold', fontsize=20)
plt.legend(labels=['Class 1', 'Class 2'])
plt.show()
```

```python
# Explode a slice of the pie chart
explode = [0.1, 0]
plt.figure(figsize=(6, 6))
df['Pclass'].value_counts().plot(kind='pie', explode=explode, autopct='%1.1f%%', colors=['#8F4700', '#F4B678'])
plt.title('Exploded Pie Chart for Passenger Class Distribution', weight='bold', fontsize=20)
plt.legend(labels=['Class 1', 'Class 2'])
plt.show()
```

```python
# Pie chart with percentage labels
plt.figure(figsize=(6, 6))
df['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#8481DD', '#8A8D90'])
plt.title('Pie Chart with Percentage Labels', weight='bold', fontsize=20)
plt.legend(labels=['Class 1', 'Class 2'])
plt.show()
```
