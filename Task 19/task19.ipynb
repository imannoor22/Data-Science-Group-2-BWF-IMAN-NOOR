# **Statistical plotting with Seaborn (regression plots, distribution plots)**
Seaborn is a powerful Python library for creating informative and attractive statistical graphics. Here's a brief overview of some key plots:

#### Regression Plots
- **Linear Regression Plot (`sns.regplot`)**: Displays a scatter plot with a fitted linear regression line for two variables.
- **Multiple Linear Regression Plot (`sns.lmplot`)**: Extends `regplot` to multiple subplots, useful for facet grids based on categorical data.
- **Residual Plot (`sns.residplot`)**: Shows the residuals (differences between observed and predicted values) of a regression.

#### Distribution Plots
- **Histogram (`sns.histplot`)**: Visualizes the distribution of a dataset by grouping data into bins.
- **Kernel Density Estimate (KDE) Plot (`sns.kdeplot`)**: Estimates the probability density function of a continuous variable.
- **Distribution Plot (`sns.distplot`)**: Combines a histogram and KDE plot (deprecated, use `sns.histplot` and `sns.kdeplot` instead).

#### Categorical Plots
- **Box Plot (`sns.boxplot`)**: Displays the distribution of a dataset based on a five-number summary (minimum, first quartile, median, third quartile, and maximum).
- **Violin Plot (`sns.violinplot`)**: Combines a box plot and KDE plot to show data distribution and density.

#### Pair Plot
- **Pair Plot (`sns.pairplot`)**: Creates a grid of scatter plots for pairwise relationships in a dataset, including histograms on the diagonal.

#### Heatmap
- **Heatmap (`sns.heatmap`)**: Displays a matrix of data as a color-encoded grid.

#### Joint Plot
- **Joint Plot (`sns.jointplot`)**: Shows the relationship between two variables along with their marginal distributions.

#### Facet Grid
- **Facet Grid (`sns.FacetGrid`)**: Enables the creation of multiple subplots based on categorical variables, useful for visualizing complex data relationships.


## Importing necessary tools

```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

# ðŸŒ¸ðŸ“Š Iris Dataset ðŸ“ˆðŸŒº

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.set_index('Id', inplace=True)
df.head()
```

## Linear Regression Plot (`sns.regplot`)

```python
sns.regplot(x='PetalLengthCm', y='SepalLengthCm', data=df);
```

## Multiple Linear Regression Plot (`sns.lmplot`)

```python
sns.lmplot(x='PetalLengthCm', y='SepalLengthCm', data=df);
```

```python
sns.lmplot(x='PetalLengthCm', y='SepalLengthCm', data=df, x_estimator=np.mean);
```

## Residual Plot (`sns.residplot`)

```python
sns.residplot(data=df, x="SepalWidthCm", y="PetalWidthCm");
```

## Histogram (`sns.histplot`)

```python
sns.histplot(data=df, x="SepalLengthCm");
```

```python
sns.histplot(data=df, x="SepalWidthCm");
```

## Kernel Density Estimate (KDE) Plot (`sns.kdeplot`)

```python
sns.kdeplot(data=df);
```

```python
sns.kdeplot(data=df, x='SepalWidthCm', hue='Species', multiple='stack');
```

```python
sns.kdeplot(data=df, x='SepalLengthCm', hue='Species', fill=True, common_norm=False,
           palette='crest', alpha=0.5, linewidth=0);
```

```python
sns.kdeplot(
    data=df, x='SepalLengthCm', y='PetalLengthCm', fill=True
);
```

```python
sns.kdeplot(data=df, x='SepalLengthCm', y='PetalLengthCm',
    fill=True, thresh=0, levels=100, cmap='mako');
```

## Distribution Plot (`sns.distplot`)
Instead of distplot, we'll use displot

```python
sns.displot(data=df, x='SepalWidthCm', kde=True);
```

## Box Plot (`sns.boxplot`)

```python
sns.boxplot(data=df, x='SepalWidthCm', y='Species');
```

```python
sns.boxplot(
    data=df, x='Species', y='SepalWidthCm',
    notch=True, showcaps=False,
    flierprops={'marker': 'x'},
    boxprops={'facecolor': (.3, .5, .7, .5)},
    medianprops={'color': 'r', 'linewidth': 2},
);
```

## Violin Plot (`sns.violinplot`)

```python
sns.violinplot(x=df['SepalWidthCm']);
```

```python
sns.violinplot(x=df['SepalWidthCm'], y=df['Species']);
```

```python
sns.violinplot(data=df, x='Species', y='SepalWidthCm', split=True, inner='quart');
```

## Pair Plot (`sns.pairplot`) 

```python
custom_palette = sns.color_palette(["#FF6F61", "#6B5B95", "#88B04B"])
sns.pairplot(df, hue='Species', palette=custom_palette, height=2);
```

## Heatmap (`sns.heatmap`)

```python
# Finding correlation matrix
corr = df.corr()
corr
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.title('Heatmap of Iris Dataset Features', weight='bold', fontsize=18)
plt.show();
```

## Joint Plot (`sns.jointplot`)

```python
sns.jointplot(data=df, x='SepalWidthCm', y='PetalWidthCm');
```

```python
sns.jointplot(data=df, x='SepalWidthCm', y='PetalWidthCm', kind='reg');
```

```python
sns.jointplot(data=df, x='SepalWidthCm', y='PetalWidthCm', hue='Species');
```

```python
sns.jointplot(data=df, x='SepalWidthCm', y='PetalWidthCm', kind='hex');
```

```python
sns.jointplot(data=df, x='SepalWidthCm', y='PetalWidthCm', hue='Species', kind='kde');
```

## Facet Grid (`sns.FacetGrid`)

```python
g = sns.FacetGrid(df, col='Species')
g.map(sns.scatterplot, 'SepalLengthCm', 'SepalWidthCm')
g.set_axis_labels('Sepal Length (cm)', 'Sepal Width (cm)')
g.add_legend()
plt.show();
```
