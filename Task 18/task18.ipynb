# **Introduction to Seaborn (visualization with Seaborn)**

- Seaborn is a library for making `statistical graphics` in Python. 
- It builds on top of matplotlib and integrates closely with pandas data structures.
- Seaborn helps you `explore` and `understand your data`.
- Its plotting functions operate on `dataframes` and `arrays` containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots.

## Importing necessary tools

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Scatter plot

```python
sns.set_theme(style="whitegrid")
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=df,
    x="sugars", y="rating",
    hue="carbo", size="protein",
    palette=cmap, sizes=(10, 200),
)
g.set(xscale="linear", yscale="linear")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True);
```

```python
sns.scatterplot(data=df, x="sugars", y="rating", hue="fat", palette="deep");
```

# Customization

## Scatter plot

```python
sns.set_style("ticks")
sns.set_palette('pastel')
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=df,
    x="sugars", y="rating",
    hue="carbo", size="protein",
    sizes=(10, 200)
)
g.set(xscale="linear", yscale="linear")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
g.fig.suptitle("Relationship Between Sugars, Rating, Carbo, and Protein", weight="bold", fontsize=16);
```

## Line plot

```python
sns.set_style("whitegrid") 
sns.set_palette('pastel')
line_plot = sns.lineplot(x="sugars", y="calories", data=df);
line_plot.set_title("Relationship Between Sugars and Calories", weight="bold", fontsize=16)  
line_plot.set_xlabel("Sugars")  
line_plot.set_ylabel("Calories");
```

## Bar plot

```python
sns.set_style("darkgrid") 
sns.set_palette('colorblind')
bar_plot = sns.barplot(data=df, x="mfr", y="rating");
bar_plot.set_title("Relationship Between mfr and ratings", weight="bold", fontsize=16)  
bar_plot.set_xlabel("MFR")  
bar_plot.set_ylabel("Rating");
```

## Count Plot

```python
sns.set_style("darkgrid") 
sns.set_palette('deep')
count_bar = sns.countplot(data=df, x="mfr", hue="type");
count_bar.set_title("MFR Distribution", weight="bold", fontsize=16)  
count_bar.set_xlabel("MFR")  
count_bar.set_ylabel("Type");
```

## Box plot

```python
sns.set_style("whitegrid") 
boxplot = sns.boxplot(data=df, x="mfr", y="rating", color=".8", linewidth=.75)
for artist in boxplot.artists:
    artist.set_edgecolor('#137')
    artist.set_linewidth(0.75)
boxplot.set_title("Relationship Between mfr and ratings", weight="bold", fontsize=16)  
boxplot.set_xlabel("MFR")  
boxplot.set_ylabel("Rating");
```
